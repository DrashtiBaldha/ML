```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('lotus.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#box blur
box=cv2.blur(img,(30,30))
plt.subplot(1,2,1),plt.imshow(img)
plt.subplot(1,2,2),plt.imshow(box)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('pic.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#median blur
med=cv2.medianBlur(img,151)
plt.subplot(1,2,1),plt.imshow(img)
plt.subplot(1,2,2),plt.imshow(med)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('imgg.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

box=cv2.blur(img,(30,30))
gaus=cv2.GaussianBlur(img,(11,11),0)
med=cv2.medianBlur(img,151)
plt.subplot(2,2,1),plt.imshow(img),plt.title('source Image')
plt.subplot(2,2,2),plt.imshow(med),plt.title('box Image')
plt.subplot(2,2,3),plt.imshow(box),plt.title('gaussian Image')
plt.subplot(2,2,4),plt.imshow(gaus),plt.title('median Image')
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('gate.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
size=30

#horizontal blur
k1=np.zeros((size,size))
k1[int(size/2),:]=np.ones(size)
k1=k1/size
horizontal=cv2.filter2D(img,-1,k1)
plt.subplot(1,2,1),plt.imshow(img),plt.title('source image')
plt.subplot(1,2,2),plt.imshow(horizontal),plt.title('Horizontal blur')
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('gate.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
size=30

#vertical blur
k1=np.zeros((size,size))
k1[:,int(size/2)]=np.ones(size)
k1=k1/size
vertical=cv2.filter2D(img,-1,k1)
plt.subplot(1,2,1),plt.imshow(img),plt.title('source image')
plt.subplot(1,2,2),plt.imshow(vertical),plt.title('vertical blur')
```
