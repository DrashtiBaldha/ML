```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('0to9.jpg')
k1=np.ones((3,3),np.uint8)

# eration
erosion=cv2.erode(img,k1,iterations=3)

#dilation
dilation=cv2.dilate(img,k1,iterations=3)

plt.subplot(1,3,1),plt.imshow(img,cmap='gray'),plt.title('Source')
plt.subplot(1,3,2),plt.imshow(erosion,cmap='gray'),plt.title('Erosion')
plt.subplot(1,3,3),plt.imshow(dilation,cmap='gray'),plt.title('Delation')
```
